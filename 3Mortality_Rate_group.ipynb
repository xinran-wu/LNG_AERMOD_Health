## Script for Mortality Rate categorized by group

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from shapely.geometry import Point, Polygon
from shapely.ops import unary_union
```

```python
d1 = pd.read_csv('siteA_output.csv')
d2 = pd.read_csv('siteB_output.csv')
d3 = pd.read_csv('siteC_output.csv')
d4 = pd.read_csv('siteD_output.csv')
```

#### [1] SiteA

```python
# Create bins and labels
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
labels = ['0-20%', '20-40%', '40-60%', '60-80%', '80-100%']

# Create a new column in your DataFrame for the percentage group
d1['POC_Percent_Group'] = pd.cut(d1['PEOPCOLORP'], bins=bins, labels=labels, include_lowest=True)

sns.set_context("notebook")

plt.figure(figsize=(8, 6),dpi=300)

# Create box plot with customized colors and other elements
sns.boxplot(x='POC_Percent_Group', y='Per_Cap_Mort', data=d1, color='white',width=0.5,
                   #   patch_artist=True,  # Color the boxes
                      boxprops=dict(facecolor='white', edgecolor='dimgrey', linewidth=1),
                      whiskerprops=dict(color='dimgrey', linewidth=1),
                      capprops=dict(color='dimgrey', linewidth=1),
                      flierprops=dict(marker='o', color='dimgrey', markersize=6, markeredgewidth=1), # whis=6, this command adjust the whisker
                      medianprops=dict(color='grey', linewidth=2))

# Calculate means for each group
means1 = d1.groupby('POC_Percent_Group')['Per_Cap_Mort'].mean().values

# Add mean line on top of boxplot
plt.plot(range(len(means1)), means1, color='#66b8ec', marker='o', linestyle='-', linewidth=2, label='Mean')


# Adding title and labels
# Customize plot
plt.xlabel('People of Color Percentage Group', fontsize=11)
plt.ylabel('$NO_2$-attributable Per Capita Mortality (deaths per 100,000 people)', fontsize=11)
#plt.title('NO2-attributable Mortality by Percent People-of-Color Group', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(-0.1,45)

plt.tight_layout()
plt.show()
```
